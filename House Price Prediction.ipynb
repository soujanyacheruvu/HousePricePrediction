```python
# Define the function to treat outliers
def treat_outliers(df, numeric_columns, upper_cap=0.90, lower_cap=0.10):
    for col in numeric_columns:
        # Calculate the quartiles and IQR
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        upper_bound = q3 + (1.5 * iqr)
        lower_bound = q1 - (1.5 * iqr)
        
    # Replace outliers with upper or lower cap
        df[col] = np.where(df[col] > upper_bound, upper_cap, df[col])
        df[col] = np.where(df[col] < lower_bound, lower_cap, df[col])
    
    return df
```

```python
# Univariate Analysis
# Description, histogram and boxplot of numerical variable

def univariateAnalysis_numeric(df, nbins):   
    numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
    for column in numerical_columns:
        print("Description of " + column)
        print("----------------------------------------------------------------------------")
        print(df[column].describe())
        print('\n')
        
        plt.figure()
        print("Distribution of " + column)
        print("----------------------------------------------------------------------------")
        sns.distplot(df[column], kde=False, color='g', bins=nbins)
        plt.show()
        
        plt.figure()
        print("BoxPlot of " + column)
        print("----------------------------------------------------------------------------")
        sns.boxplot(x=df[column])
        plt.show()
```

```python
# Select categorical columns in the DataFrame
df_cat = df.select_dtypes(['object', 'category'])

def univariateAnalysis_category(cat_column):
    # Print percentage details of the categorical column
    print("Percentage Details of " + cat_column)
    print("----------------------------------------------------------------")
    
    # Calculate the percentage of each category and sort by index
    percentages = (df_cat[cat_column].value_counts(normalize=True).sort_index()) * 100
    print(round(percentages, 2))
    print("\n")
    
    # Plot the frequency distribution of the categorical column
    plt.figure()
    df_cat[cat_column].value_counts().sort_index().plot.bar(title="Frequency Distribution of " + cat_column)
    plt.show()
    print("\n")

# Get a list of categorical column names
lstcatcolumns = list(df_cat.columns.values)
```
